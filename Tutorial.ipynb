# Whats actually happening

This is a time forecasting model, meaning it is a machine learning algorthim that forecasts values based on time. There are many time forecasting algorithms and we have used the basic one called ARIMA, which is short for Auto Regressive Integrated Moving Average. The details of how the model is configured and how to choose the parameters in given in **Stock_Predictor\Models\ARIMA.ipynb**

The main reference of ARIMA is from SARIMA, which is short for Seasonal correlation for ARIMA. It can be found on 
https://towardsdatascience.com/how-to-forecast-sales-with-python-using-sarima-model-ba600992fa7d
